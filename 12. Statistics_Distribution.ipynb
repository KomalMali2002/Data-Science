# @ Power Law Distribution 

 - It is following the 80-20 rule
 - The distribution which follows the power law we called it as pareto disctribution
 - We have two distribution parameter
 - (alpha --- scale/ mean)
 -  (Xm ---> SD)
 - CDF = p (X<x)
 - CCDF = P (X>x) ----- complement cumulative distribution function


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from statsmodels import robust
```

```python
data = [23,45,67,89,90,101,102]
```

```python
stats.probplot(data, dist='pareto', sparams=(2.5,), plot=plt)
plt.show()
```

```python
stats.probplot(data, dist='pareto', sparams=(9.5,), plot=plt)
plt.show()
```

```python
data=np.sort([23,45,67,89,90,101,102])
cdf= np.arange(1, len(data) + 1) / len(data)
ccdf=1-cdf
plt.loglog(data, ccdf, marker='o', linestyle='-', color='b')
plt.show()
```

```python
data = [23, 45, 67, 89, 90, 101, 102]
sorted_data = np.sort(data)
cdf = np.arange(1, len(sorted_data) + 1) / len(sorted_data)
ccdf = 1 - cdf
plt.figure()
plt.step(sorted_data, ccdf, where='post', label='Empirical CCDF')
plt.xlabel('x')
plt.ylabel('P(X > x)')
plt.title('Empirical CCDF')
plt.grid(True)
plt.legend()
plt.show()
```

```python
data = [23, 45, 67, 89, 90, 101, 102]
sorted_data = np.sort(data)
cdf = np.arange(1, len(sorted_data) + 1) / len(sorted_data)
ccdf = 1 - cdf
nonzero_indices = ccdf > 0
x = sorted_data[nonzero_indices]
y = ccdf[nonzero_indices]
plt.loglog(x, y, marker='o', linestyle='-', label='Empirical CCDF')
plt.xlabel('Data (log scale)')
plt.ylabel('CCDF (log scale)')
plt.title('Log-Log Plot of Empirical CCDF')
plt.legend()
plt.grid(True, which='both', linestyle='--')
plt.show()
```

# @ boxcox transformation :
- X~pareto distribution(alpha,Xm)

  x1,x2,x3-----xn------pareto 
- Y~GD(mean,sigma)

  y1,y2,y3-------yn------gaussian

- import scipy.stats as stats
- Syntax:stats.boxcox(x)------>lambda

- y1=((x1^lambda)-1)/lambda when lambda>0

- log(x)-----lambda=0

- when the lambda=0 then x follows the log normal distribution
