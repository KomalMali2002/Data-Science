```python
df1=pd.read_csv("D:\\Download\\iris.csv")
df1
```

```python
df1.plot(kind="scatter", x="sepal.length", y="sepal.width")
plt.show()
```

- sepal length in this dataset varies from 4 to 8
- sepal width is this dataset varies from 2 to 4.5

```python
sns.set_style("whitegrid")
sns.FacetGrid(df1,hue='variety',height=4).map(plt.scatter ,'sepal.length',"sepal.width").add_legend();
plt.show()
```

- iris setosa is somewhat seperable while virginica and versicolor are overlapping

```python
sns.set_style("whitegrid")
sns.FacetGrid(df1,hue='variety',height=4).map(plt.scatter ,'petal.length',"petal.width").add_legend();
plt.show()
```

 - If petal length is greater than or eqaul to 0.8 and petal length  is less than or eqaul to 2

 - If (PL>=0.8 and PL<=2) and (PW>=0.2 and PW<=0.6)-------iris setosa

 - If petal width is greater than or eqaul to 0.8 and petal width  is less than or eqaul to 2

 -  If (PL>=2.8 and PL<=3.3) and (PW>=0.8 and PW<=1.7)-------iris versicolor
 
 - If (PL>=4.5 and PL<=6.9) and (PW>=1.4 and PW<=2.6)-------iris virginica

```python
sns.pairplot(df1,hue="variety",size=3,diag_kind='kde')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(df1,hue='variety',height=3).map(sns.distplot,'sepal.length').add_legend();
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(df1,hue='variety',height=3).map(sns.distplot,'sepal.width').add_legend();
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(df1,hue='variety',height=3).map(sns.distplot,'petal.length').add_legend();
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(df1,hue='variety',height=3).map(sns.distplot,'petal.width').add_legend();
plt.show()
```

-  PL far better than PW and it is better then SL and SL is better than SW
-  PL>>PW>>SL>>SW

```python
sns.boxplot(x="variety", y="sepal.length", data=df1)
plt.show()
```

- (SL>=4.3 and SL<=5.8)----for setosa
- (SL>=4.9 and SL<=7.0)-----for versicolor
- (SL>=5.6 and SL<=7.9)-----for virginica-----it has outliers

```python
sns.boxplot(x="variety", y="sepal.width", data=df1)
plt.show()
```

- (SW>=2.9 and SW<=4.3)----for setosa-----it has outliers
- (SW>=2.0 and SW<=3.4)---for versicolor
- (SW>=2.5 and SW<=3.6)-----for virginica----it has outliers

```python
sns.boxplot(x="variety", y="petal.length", data=df1)
plt.show()
```

- (PL>=1.2 and PL<=1.7)----for setosa----it has outliers
- (PL>=3.3 and PL<=5.1)-----for versicolor----it has outliers
- (PL>=4.5 and PL<=6.9)-----for virginica

```python
sns.boxplot(x="variety", y="petal.width", data=df1)
plt.show()
```
